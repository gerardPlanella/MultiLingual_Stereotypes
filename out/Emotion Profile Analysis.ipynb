```python
def find_emotions(folder_path):

#     folder_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/english/"
    # english_pretrain = pd.read_csv(file_path)

    dataframes = []

    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a CSV file
        if filename.endswith(".csv"):
            # Construct the file path
            file_path = os.path.join(folder_path, filename)
            # Load the CSV file into a DataFrame
            df = pd.read_csv(file_path)
            # Append the DataFrame to the list
            dataframes.append(df)

    # Concatenate all DataFrames into one
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Now you can work with the combined DataFrame
    combined_df.rename( columns={'Unnamed: 0':'group'}, inplace=True )
    combined_df = combined_df.iloc[:, 1:]
    return combined_df.mean(0)
```

```python
def find_race(folder_path):
    file_path = os.path.join(folder_path, "race.csv")
    df = pd.read_csv(file_path)
    df.rename(columns={'Unnamed: 0':'group'}, inplace=True)
    df = df.iloc[:, 1:]
    return df.mean(0)
    
```

```python
english_pretrain_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/english/"
french_pretrain_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/french/"
spanish_pretrain_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/spanish/"
greek_pretrain_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/greek/"
croatian_pretrain_path = "/Users/veljko/Desktop/atcs/atcs-project/MultiLingual_Stereotypes/out/pretrained_roberta/emotion_profiles/croatian/"



english_pretrain = find_emotions(english_pretrain_path)
french_pretrain = find_emotions(french_pretrain_path)
spanish_pretrain = find_emotions(spanish_pretrain_path)
greek_pretrain = find_emotions(greek_pretrain_path)
croatian_pretrain = find_emotions(croatian_pretrain_path)
```
